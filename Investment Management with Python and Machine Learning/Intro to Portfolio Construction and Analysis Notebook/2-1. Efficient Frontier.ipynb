Let's see the performance in 1996 - 2000.

```python
#Returns
er = rk.annualized_rets(ind['1996':'2000'], 12)
er.sort_values().plot(kind='bar')
```

```python
#Covariance metirc
cov= ind['1996':'2000'].cov()
cov
```

```python
#Generate portfolio
#We've add these code to our module kit file
def portfolio_return(weights, returns):
    """
    Weights -> Returnss
    """
    return weights.T @returns

def portfolio_vol(weights, covmat):
    """
    Weights -> Vol
    """
    return (weights.T @covmat @weights)**0.5
```

```python
#Ger annualized returns
l= ['Food', 'Beer', 'Smoke', 'Coal']
er[l]
```

```python
#Cov matrix
cov.loc[l,l]
```

```python
weights = np.repeat(1/4, 4) # This will generate a array of 0.25, 0.25, 0.25 ,0.25
rk.portfolio_return(weights, er[l])
```

```python
#Compute the vol of our 4 assets
rk.portfolio_vol(weights, cov.loc[l,l])
```

### 2 assets Frontier
Let's start build the frontier of 2 assets.

```python
l = ['Games', 'Fin']
n_points= 20
weights= [np.array([w, 1-w]) for w in np.linspace(0, 1, n_points)] # A sequence of weights
weights
```

```python
len(weights)
```

```python
rets = [rk.portfolio_return(w, er[l]) for w in weights]
vols = [rk.portfolio_vol(w, cov.loc[l,l]) for w in weights]
efficient_frontier = pd.DataFrame({'Return': rets, 'Vol':vols})
efficient_frontier.head()
```

```python
efficient_frontier.plot.scatter(x="Vol", y="Return")
```
