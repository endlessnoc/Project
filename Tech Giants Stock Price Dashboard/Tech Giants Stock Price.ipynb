# Tech Giants Stock Price Dashboard Project

This project involves using the Yahoo Finance API to retrieve stock market information, performing initial data processing with pandas, and generating interactive reports utilizing Tableau. We will be working with market data from 'Tech giants', including Amazon, Apple, Google, Meta, and Microsoft.  

Tableau Public Link: https://public.tableau.com/app/profile/wei.hsin.hsu/viz/TechGiantsStockPriceDashboard/Dashboard1

### 1. Import Libraries


```python
import yfinance as yf
import pandas as pd
```

### 2. Obtain data


```python
#Utilize yfinance api to access market data
def get_stock_data(ticker, start_date): 
    stock = yf.download(ticker, start=start_date).reset_index()
    return stock

company_list = ['AMZN', 'AAPL', 'GOOG', 'META', 'MSFT']
start_date = '2013-02-01'

stock_dict = {}
for ticker in company_list:
    stock_dict[ticker] = get_stock_data(ticker, start_date)


```

### 3. Data Processing


```python
for df in company_list:
    stock_dict[df]['Date']= pd.to_datetime(stock_dict[df]['Date'])
    stock_dict[df]['MA5'] = stock_dict[df]['Close'].rolling(5).mean()
    stock_dict[df]['MA20'] = stock_dict[df]['Close'].rolling(20).mean()
    stock_dict[df]['MA60'] = stock_dict[df]['Close'].rolling(60).mean()
    stock_dict[df]['MA120'] = stock_dict[df]['Close'].rolling(120).mean()
    stock_dict[df]['MA240'] = stock_dict[df]['Close'].rolling(240).mean()
```

```python
stock_dict['AAPL'].head()
```

```python
for df in company_list:
    stock_dict[df]['Prev_Close'] = stock_dict[df]['Close'].shift(1)
    stock_dict[df]['Change_Close'] = stock_dict[df]['Close'] - stock_dict[df]['Prev_Close']
    stock_dict[df]['%_Close_Change'] = stock_dict[df]['Close'].pct_change()
    stock_dict[df]['Prev_Vol'] = stock_dict[df]['Volume'].shift(1)
    stock_dict[df]['Change_Vol'] = stock_dict[df]['Volume'] - stock_dict[df]['Prev_Vol']
    stock_dict[df]['%_Vol_Change'] = stock_dict[df]['Volume'].pct_change()
```

### 4. Export Data


```python
stock_dict['AMZN'].to_csv('Amazon.csv')
stock_dict['AAPL'].to_csv('Apple.csv')
stock_dict['GOOG'].to_csv('Google.csv')
stock_dict['META'].to_csv('Meta.csv')
stock_dict['MSFT'].to_csv('Microsoft.csv')
```
